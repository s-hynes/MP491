### Lecture 20 Example 1

$$ x_{n+1} = a x_n (1-x_n) \quad , \quad x\in[0,1] \quad , \quad a=\frac{33}{10}$$

```python
from CH05_1d_maps import conv
conv(0, 33/10, 100, lambda r,x: r*x*(1-x) )
conv(23/33, 33/10, 100, lambda r,x: r*x*(1-x) )
```

```python
from CH05_1d_maps import conv, mapping
# This code cell is to show the period two orbit that happens with this value of a.

# Showing this graphically
conv(0.1, 33/10, 100, lambda r,x: r*x*(1-x) )

x_n = mapping(0.1, 33/10, 100, lambda r,x: r*x*(1-x))

# Showing this by looking at the values of x_n for the last twenty iterations (81-100)
print("Values at last twenty iterations:\n", x_n[-20:], "\n")

# It tends towards this periodic orbit sooner than this though, after about 30 iterations
# (but this depends on how many significant figures you take the values to):
for i in range(20,41):
    print("n = {0} iterations, x\u2099 = {1:.6}".format(i, x_n[i]))
```
