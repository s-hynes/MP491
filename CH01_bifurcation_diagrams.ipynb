### What is this?

This Jupyter notebook uses the function `plotter` from CH01_bifurcation.py to plot different bifurcation diagrams that come up in Chapter 1 of the course. Points plotted in red are unstable fixed points and points plotted in green are unstable (I haven't been able to get the dashed and solid lines working for this yet).

```python
from CH01_bifurcation import plotter
```

### Saddle-Node Bifurcation

```python
from CH01_bifurcation import plotter
plotter((lambda r: (lambda x: x**2 - r)), fstring="x**2-r",rmin=-0.1, rmax=5, rstep=0.01)
```

```python
from CH01_bifurcation import plotter
plotter( (lambda r: (lambda x: x**2 + r)), fstring="x**2+r",rmin=-5, rmax=0.1, rstep=0.01)
```

### Pitchfork Bifurcation

#### Subcritical Pitchfork Bifurcation

```python
from CH01_bifurcation import plotter
plotter( (lambda r: (lambda x: x**3 + r*x)), fstring="x**3+r*x",rmin=-5, rmax=5, rstep=0.01)
```

```python
from CH01_bifurcation import plotter
plotter( (lambda r: (lambda x: x**3 - r*x)), fstring="x**3-r*x",rmin=-5, rmax=5, rstep=0.01)
```

#### Supercritical Pitchfork Bifurcation

```python
from CH01_bifurcation import plotter
plotter( (lambda r: (lambda x: -x**3 + r*x)), fstring="-x**3+r*x",rmin=-5, rmax=5, rstep=0.01)
```

```python
from CH01_bifurcation import plotter
plotter( (lambda r: (lambda x: -x**3 - r*x)), fstring="-x**3-r*x",rmin=-5, rmax=5, rstep=0.01)
```

### Transcritical Bifurcation
