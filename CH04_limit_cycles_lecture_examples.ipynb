### Lecture 16 Example 1
$$
\begin{align*}
    \dot r &= r(1-r^2) \\
    \dot \theta &= 1
\end{align*}
$$
In Cartesian coordinates, these ODEs are:
$$
\begin{align*}
    \dot x &= x(1-x^2-y^2) - y\\
    \dot y &= x + y(1-x^2-y^2)
\end{align*}
$$

```python
from CH03_nonlinear import stream
from CH04_polar_coords import stream_pol
stream(lambda x,y: x*(1-x**2-y**2) - y, lambda x,y: x + y*(1-x**2-y**2), xmin=-2, xmax=2, ymin=-2, ymax=2, eq_pts=[[0,0]])
stream_pol(lambda r,theta: r*(1-r**2), lambda r,theta: 1, xmin=-2, xmax=2, ymin=-2, ymax=2, eq_pts=[[0,0]])
```
