### Problem Sheet 5 Question 1
$$x_{n+1} = 2 x_n - 2 (x_n)^2$$

<details>
<summary><strong>Fixed Points and their Stabilities</strong></summary>

$x^* = 0            \ \text{(unstable)}\\ 
x^* = \frac{1}{2}   \ \text{(stable)} $

</details>

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = 0$ is a fixed point. When $x_0 = 0$, $x_n = 0$ for all subsequent iterations n.

</details>

```python
from one_dimensional_maps import conv
conv(0, 0.5, 100, lambda r,x: 2*x - 2*(x**2))
```

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = 0$ is an unstable fixed point and $x^* = \frac{1}{2}$ is a stable fixed point. Even when $x_0$ is close to 0, $x_n \rightarrow \frac{1}{2}$ after sufficiently many iterations.

</details>

```python
from one_dimensional_maps import conv
conv(0.000000001, 0.5, 100, lambda r,x: 2*x - 2*(x**2))
```

### Problem Sheet 5 Question 3

$$x_{n+1} = (x_n)^2 - x_n - 2$$

<details>
<summary><strong>Fixed Points and their Stabilities</strong></summary>

$x^* = 1 + \sqrt{3} \approx \ \ \ 2.73 \quad \text{(unstable)}\\ 
x^*  = 1 - \sqrt{3} \approx -0.73  \quad \text{(unstable)} $

</details>

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = 1 + \sqrt{3}$ is an unstable fixed point. When $x_0 = 1 + \sqrt{3}$, $x_n = 1 + \sqrt{3}$ for all subsequent iterations n, but if $x_0$ is close to $1 + \sqrt{3}$ but not quite equal, $x_n$ eventually blows up. It doesn't settle down to the fixed point.

</details>

```python
from one_dimensional_maps import conv
import numpy as np
import warnings
warnings.filterwarnings("ignore", ".*scalar power.*")
warnings.filterwarnings("ignore", ".*scalar subtract.*")

# With x_0 = 1 + sqrt(3), x_n remains the same for subsequent iterations
conv(1+np.sqrt(3), 0.5, 500, lambda r,x: x**2 - x - 2)

# Changing x_0 by even a small amount leads to x_n eventually blowing up.
# The convergence part of this code doesn't work for cases like this, so just ignore that output.
conv(1+np.sqrt(3)+0.001, 0.5, 500, lambda r,x: x**2 - x - 2)

# Putting x_0 farther away from this unstable fixed point. Again, x_n blows up eventually.
conv(2+np.sqrt(3), 0.5, 500, lambda r,x: x**2 - x - 2)
```

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = 1 - \sqrt{3}$ is an unstable fixed point. When $x_0 = 1 - \sqrt{3}$, $x_n = 1 - \sqrt{3}$ for all subsequent iterations n, but if $x_0$ is close to $1 - \sqrt{3}$ but not quite equal, $x_n$ eventually blows up. It doesn't settle down to the fixed point.

</details>

```python
from one_dimensional_maps import conv
import numpy as np
import warnings
warnings.filterwarnings("ignore", ".*scalar.*")

# With x_0 = 1 - sqrt(3), x_n remains the same for subsequent iterations
conv(1-np.sqrt(3), 0.5, 500, lambda r,x: x**2 - x - 2)

# Changing x_0 by even a small amount leads to x_n eventually blowing up.
# The convergence part of this code doesn't work for cases like this, so just ignore that output.
conv(1-np.sqrt(3)+0.001, 0.5, 500, lambda r,x: x**2 - x - 2)

# Putting x_0 farther away from this unstable fixed point. Again, x_n blows up eventually.
conv(0.5, 0.5, 500, lambda r,x: x**2 - x - 2)
```

### Problem Sheet 5 Question 4

$$x_{n+1} = x_n (x_n - 1)(x_n + 1)$$

<details>
<summary><strong>Fixed Points and their Stabilities</strong></summary>

$$
\begin{align}
x^* &=  \sqrt{2}   \quad & &\text{(unstable)} \nonumber\\
x^* &=  0          \quad & &\text{(linearisation can't conclude anything about its stability)} \nonumber\\ 
x^* &= -\sqrt{2}   \quad & &\text{(unstable)} \nonumber
\end{align}
$$

</details>

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = \sqrt{2}$ is an unstable fixed point. When $x_0 = \sqrt{2}$, $x_n = \sqrt{2}$ for all subsequent iterations $n$, but if $x_0$ is close to $\sqrt{2}$ but not quite equal, $x_n$ doesn't settle down to the fixed point.

</details>

```python
from one_dimensional_maps import conv
import numpy as np
import warnings
warnings.filterwarnings("ignore", ".*scalar.*")
conv(np.sqrt(2), 0.5, 150, lambda r,x: x*(x-1)*(x+1))
conv(np.sqrt(2)-0.5, 0.5, 150, lambda r,x: x*(x-1)*(x+1))
```

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = -\sqrt{2}$ is an unstable fixed point. When $x_0 = -\sqrt{2}$, $x_n = -\sqrt{2}$ for all subsequent iterations $n$, but if $x_0$ is close to $-\sqrt{2}$ but not quite equal, $x_n$ doesn't settle down to the fixed point.

</details>

```python
from one_dimensional_maps import conv
import numpy as np
import warnings
warnings.filterwarnings("ignore", ".*scalar.*")
conv(-np.sqrt(2), 0.5, 150, lambda r,x: x*(x-1)*(x+1))
conv(-np.sqrt(2)+0.5, 0.5, 150, lambda r,x: x*(x-1)*(x+1))
```

<details>
<summary><strong>Spoiler</strong></summary>

The code below shows that $x^* = 0$ is a fixed point. When $x_0 = 0$, $x_n = 0$ for all subsequent iterations $n$, but if $x_0$ is close to $0$ but not quite equal, $x_n$ doesn't settle down to the fixed point. There seems to instead be a oscillatory behaviour about $x^*=0$.

</details>

```python
from one_dimensional_maps import conv
conv(0, 0.5, 150, lambda r,x: x*(x-1)*(x+1))
conv(0.5, 0.5, 150, lambda r,x: x*(x-1)*(x+1))
```
